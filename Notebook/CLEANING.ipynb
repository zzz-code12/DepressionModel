Degree

```python
print("=" * 50)
print("DEGREE CHECK")
print("=" * 50)
print(df['Degree'].value_counts())
print(f"\nUnique values: {df['Degree'].unique()}")
print(f"Missing values: {df['Degree'].isnull().sum()}")
```

```python
print("=" * 50)
print("DEGREE CLEANING")
print("=" * 50)

# Step 1 - Remove quotes
df['Degree'] = df['Degree'].str.strip("'")

# Step 2 - Drop Others
df = df[df['Degree'] != 'Others']
print(f"Rows after dropping Others: {len(df)}")

# Step 3 - Group into categories
undergraduate = [
    'BA', 'BSc', 'BCA', 'B.Tech', 'B.Com',
    'B.Ed', 'B.Arch', 'B.Pharm', 'BBA',
    'BHM', 'BE', 'LLB', 'MBBS'
]

postgraduate = [
    'MSc', 'M.Tech', 'MCA', 'M.Com', 'M.Ed',
    'M.Pharm', 'MBA', 'MD', 'LLM', 'MHM', 'ME',
    'MA'  # ← added here!
]

other = [
    'Class 12', 'PhD'
]

# Step 4 - Map to categories
def categorize_degree(degree):
    if degree in undergraduate:
        return 1
    elif degree in postgraduate:
        return 2
    elif degree in other:
        return 3
    else:
        return None

df['Degree'] = df['Degree'].apply(categorize_degree)

# Step 5 - Drop any None values
df = df.dropna(subset=['Degree'])

print("\nAFTER CLEANING:")
print(df['Degree'].value_counts())
print(f"\nRows remaining: {len(df)}")
```
